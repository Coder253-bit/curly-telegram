This means that higher the number of people in the client's social surrounding that defaulted on 30 days past due, the higher the number of people that defaulted past 60 days in the client's social surroundings

#### The same positive correlation is seen between credit to income ratio and amount of goods price. This also makes sense as the higher the amount of goods price the bank is required to give loan towards, the higher is the probability to default on it.
#### There is a negative correlation between region population and region rating. This means that the more populated a region where the client lives, the lower is the rating of the region. This is likely because as the population of a region grows, the higher the number of people that apply for loans. And this leads to a higher number of defaulters from the region, which eventually leads to a lower rating for the region.
